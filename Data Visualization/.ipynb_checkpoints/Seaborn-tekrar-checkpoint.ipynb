# Distribution plots

Dağılım Grafikleri



Verileri analiz etmeye veya modellemeye yönelik her türlü çabanın ilk adımı değişkenlerin nasıl dağıldığını anlamak olmalıdır. Dağılım görselleştirme teknikleri birçok önemli soruya hızlı yanıtlar sağlayabilir. Gözlemler hangi aralığı kapsıyor? Merkezi eğilimleri nedir? Tek bir yöne doğru aşırı derecede çarpık mı? Çift modalite için kanıt var mı? Önemli aykırı değerler var mı? Bu soruların yanıtları diğer değişkenlerle tanımlanan alt kümeler arasında farklılık gösteriyor mu?

## sns.displot()

Bir dağılımı görselleştirmek için belki de en yaygın yaklaşım histogramdır. 

histplot() ile aynı temel kodu kullanan displot() içindeki varsayılan yaklaşımdır. Histogram, veri değişkenini temsil eden eksenin bir dizi ayrık kutuya bölündüğü ve her kutuya düşen gözlem sayısının ilgili çubuğun yüksekliği kullanılarak gösterildiği bir çubuk grafiğidir:

```python
penguins = sns.load_dataset("penguins")
penguins.sample(10)
```

```python
sns.displot(data = penguins, x = "flipper_length_mm")

plt.show()
#  en yaygın palet uzunluğunun yaklaşık 195 mm olduğunu görebiliyoruz
```

```python
sns.displot(data = penguins, x = "flipper_length_mm", bins = 25, kde = True)
# bins : kaç adet bar olucağını belirler
# kde : True ile kernel yoğunluk çizgisini gösterdi
plt.show()
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")
# element : şeffaflaştırdı
plt.show()
```

```python
sns.displot(penguins, x="flipper_length_mm", col="sex", kde = True)
# col : verilen sütuna göre hue ve facegrid gibi gösterdi

plt.show()
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde")

plt.show()
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill = True);
# fill : yoğunluk çizgi gruplarının içini renkle doldurdu
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="ecdf");

# güçlü bir perspektif sunabilir
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde");
```

## sns.hisplot()

```python
sns.histplot(data=penguins, x="flipper_length_mm", kde = True, bins = 15);
```

## sns.rugplot() 

Bireysel gözlemlerin konumunu göze batmayan bir şekilde göstererek diğer grafikleri tamamlamayı amaçlamaktadır.

```python
tips = sns.load_dataset("tips")
sns.displot(data=tips, x="total_bill", kind = "kde")
sns.rugplot(data=tips, x="total_bill");
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
sns.rugplot(data=tips, x="total_bill", y="tip", hue="time", height = .05)
```

## sns.joinplot()

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm");
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species");
```

## sns.pairplot()

Bir veri kümesindeki ikili ilişkileri çizin.

```python
sns.pairplot(penguins, hue = "species");
```

# Relational plots

İlişkisel Grafikler

Farklı özelliklerin birbiriyle olan ilişkisini(korelasyon) görüntülemek için kullanılır

## sns.scatterplot()

```python
tips = sns.load_dataset("tips")
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "sex", palette = ("blue", "red"), 
                style = "smoker", alpha = 0.5, s = 50, size = "sex")

#palette ile noktaların renklerini ayarladık
# hue : renkledirerek gruplar
# style : işaretle gruplar
#alpha : yoğunluk/ markerlerın transparanlığı
# s ile markerklerın boyutlarıyla oynayabiliriz
plt.show()
```

```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip", size = "sex")

#size : markerlerı büyükllü küçüklü grupladı
plt.show()
```

## sns.relplot()

```python
sns.relplot(
    data = tips,
    x = "total_bill", y = "tip", hue = "smoker", col = "time");
```
