# Figure and Axes renklendirme

```python
plt.figure(facecolor = "yellow", figsize = (10, 5)) # figure rengini ayarladık

ax = plt.axes()
ax.set_facecolor("xkcd:mint green") # axes rengini ayarladık

plt.plot(age, salary, "b:", label = "Salary") 
plt.plot(age, salary2, "r-", label = "Salary-2")

plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc = "best")
plt.title("SALARY BY AGE");
```

```python
plt.figure(facecolor = "yellow", figsize = (10, 5))

ax = plt.axes()
ax.set_facecolor("xkcd:mint green")

plt.plot(age, salary, "b-", label = "Salary", linewidth = 4)  # linewidth = lw şeklinde yazılabilir
plt.plot(age, salary2, "r", label = "Salary-2", linestyle = ":") # linestyle = ls şeklinde yazılabilir

plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(loc = "best")
plt.title("SALARY BY AGE");
```

# Functional Method vs Object Oriented Methods

## Functional Method

```python
plt.style.use("seaborn-darkgrid")

plt.plot(age, salary)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("SALARY BY AGE")

plt.grid(color = "purple") # style'ı gridli olanlarda gridleri modifiye edebiliriz

plt.show()
```

## OOM

```python
fig, ax = plt.subplots()

ax.plot(age, salary, "r")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("SALARY BY AGE")

plt.show()
```

```python
# functional method : multi plot

plt.subplot(2, 1, 1)
plt.plot(age, salary, "r")

plt.subplot(2, 1, 2)
plt.plot(age, salary2, "b");
```

```python
# OOM : multi plot

fig, ax = plt.subplots(2, 1, figsize = (5, 5), facecolor = "yellow")

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary2, "b")

plt.tight_layout() # axes'ler arasını otomatik düzenler(grafiklerin birbiri içine geçmesini engeller)
plt.show()
```

```python
plt.subplot(2, 3, 1)
plt.plot(age, salary, "r")

plt.subplot(2, 3, 2)
plt.plot(age, salary2, "b")

plt.subplot(2, 3, 3)
plt.plot(age, salary, "black")

plt.subplot(2, 3, 4)
plt.plot(age, salary2, "purple")

plt.subplot(2, 3, 5)
plt.plot(age, salary, "y")

plt.subplot(2, 3, 6)
plt.plot(age, salary2, "w")

plt.tight_layout()
plt.show()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault) # grafiklerdeki tüm modfikasyonları default değerlere döndürdük
```

```python
fig, ax = plt.subplots(2, 3)

ax[0, 0].plot(age, salary, "red")
ax[0][1].plot(age, salary2, "black")
ax[0, 2].plot(age, salary, "brown")
ax[1, 0].plot(age, salary2, "pink")
ax[1, 1].plot(age, salary, "blue")
ax[1, 2].plot(age, salary2, "purple")

plt.tight_layout()
plt.show()
```

```python
x = np.arange(10)

fig, ax = plt.subplots(2, 2, figsize = (12,6))

plt.suptitle("LineStyles & Markers & Markersize & Color", fontsize = 20, y = 0.93)

ax[0, 0].plot(x, "r", marker= "+", markersize = 5)
ax[0, 1].plot(x-2, "y", marker= "o", ms = 5)

ax[1, 0].plot(x-3, "black", marker= "s", ms = 5, lw = 3)
ax[1, 1].plot(x-4, "g", marker= "1", ms = 5)

plt.show()
```

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)



fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))

ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)

plt.tight_layout()
plt.show()
```

# plt.pie()

pie chart

```python
x = np.array([100, 100, 100, 100])
my_labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.figure(figsize = (10, 8))
plt.pie(x, labels = my_labels, labeldistance = 0.5, autopct = "%.2f")

#labels diye tanımlamak label etikerlerini tanımlamak gerekir
#labeldistance : label yazılarının yerini ayarladık
#autopct : oranlarını gösterdi

plt.show()
```

```python
z = np.array([16, 21, 45, 19])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.figure(figsize = (7, 7))
plt.pie(z, labels = mylabels, explode = myexplode, labeldistance = 0.75, autopct = "%.1f", startangle = 45)
#startangle : pie'yi kendi ekseni etrafında döndürür
#explode : dilimi pastadan ayrık göstermek için

plt.show()
```

```python
colors1 = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explodes = [0.2, 0.1, 0.05, 0.15]

plt.figure(figsize = (10, 8))
plt.pie(z, labels = mylabels, explode = explodes, labeldistance = 0.65, 
        autopct = "%.1f", pctdistance = 0.3, colors = colors1, shadow = True)
#pstdistance ile autopct' nin yerlerini ayarladık
#shadow : gölgelendirme yaptı

plt.legend()
plt.show()
```
