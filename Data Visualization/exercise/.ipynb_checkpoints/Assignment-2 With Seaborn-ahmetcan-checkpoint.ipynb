## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
sns.catplot(data = kill, x = "gender", y = "age", kind = "swarm", hue = "manner_of_death");
```

```python
sns.swarmplot(data = kill, x= "gender", y = "age", hue = "manner_of_death");
```

## 2: Plot 12 most dangerous cities

```python
kill.city.value_counts()[0:12]
```

```python
np.array(kill.city.value_counts()[0:12]).astype(int)
```

```python
mdc = kill.city.value_counts()[0:12]
mdc.index
```

```python
sns.countplot(x = mdc)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.barplot(x = mdc.index, y = mdc.values)

plt.title("Most Dangerous Cities", color = "red")

plt.xticks(rotation=45, color = "red");
```

```python
# with seaborn
```

```python
mdc
```

```python
mdc.index
```

```python
fig, ax = plt.subplots(figsize = (5, 3))

sns.barplot(x = mdc.index, y = mdc.values)

plt.title("Most Dangerous Cities", color = "blue")

plt.xticks(rotation=45, size = 8)
plt.yticks(size = 8)
plt.text(5, -13, "city", size = 8)
plt.text(-2, 15, "count", size = 8, rotation = 90);
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
mds = kill.state.value_counts()[0:20]
mds
```

```python
fig, ax = plt.subplots(figsize = (6, 4))

sns.barplot(x = mds.index, y = mds.values)

plt.title("Most Dangerous States", color = "blue")

plt.xticks(rotation=45, size = 6)
plt.yticks(size = 6);
```

```python
# with seaborn barplot
```
