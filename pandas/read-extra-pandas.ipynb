# sql connection

```python
!pip install sqlalchemy
from sqlalchemy import create_engine
```

```python
temp_db = create_engine("sqlite:///:memory:") # geçiçi olarak bir veritabanı oluşturduk
```

```python
temp_db.table_names() # databasedeki table isimlerini sorguladık
```

```python
tables = pd.read_html("https://en.wikipedia.org/wiki/World_population")
```

```python
tables[5]
```

```python
tables[5].to_sql(name="populations", con=temp_db)
```

```python
temp_db.table_names()
```

## read

```python
df = pd.read_sql(sql = "populations", con = temp_db, index_col="index")
df
```

```python
pd.read_sql_query(sql = "SELECT Country, Population FROM populations", con=temp_db) # sorgulu çağırma
```
